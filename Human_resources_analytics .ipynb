6. (i)Encoding the Independent variable i.e., Departments and concatinating with dataset

```python
dept=pd.get_dummies(df["Departments "],drop_first=True)
df=pd.concat([df,dept],axis=1)
```

6. (ii)Encoding the Independent variable i.e., salary and concatinating with dataset

```python
sal=pd.get_dummies(df["salary"],drop_first=True)
df=pd.concat([df,sal],axis=1)
```

```python
df
```

Departments and salary columns are encoded so we can remove those columns

```python
df.drop(["Departments "],axis=1,inplace=True)
df.drop(["salary"],axis=1,inplace=True)
```
